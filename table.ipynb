# Genrate data 

```python
import turtle as t
import random as rd
import pandas as pd
import csv
import os


for i in range (1,100):
    #Clear the screen
    t.clear()

    # Reset position and heading
    t.penup()
    t.home()
    t.pendown()



    #Move the screen so 0,0 is in the bottom left, and zoom out
    t.setup(800, 800)
    t.setworldcoordinates(0, 0, 1000, 1000)


    # Generate two random integers between 1 and 10
    height = rd.randint(1,11)
    width = rd.randint(1,11)

    #add two zeros to the end of the integers, unless they are 10
    if height == 10:
        height = str(height) + '0'
        height = int(height)
    else:
        height = str(height) + '00'
        height = int(height)
    if width == 10:
        width = str(width) + '0'
        width = int(width)
    else:
        width = str(width) + '00'
        width = int(width)

    # Set turtle speed
    t.speed(20)

    # If the turtles position is less than width, run the loop
    while t.xcor() < width:
        t.left(90)
        t.forward(height)
        t.penup()
        t.backward(height)
        t.right(90)
        t.forward(100)
        t.pendown()
    # Finish the grid for height
    t.left(90)
    t.forward(height)
    t.backward(height)

    # Reset the turtles position
    t.penup()
    t.goto(0,0)
    t.pendown()

    #Setup the bottom line
    t.right(90)
    t.forward(width)
    t.penup()
    t.backward(width)
    t.left(90)

    # If the turtles position is less than height, run the loop
    while t.ycor() < height:
        t.penup()
        t.forward(100)
        t.pendown()
        t.right(90)
        t.forward(width)
        t.penup()
        t.backward(width)
        t.left(90)

    # Move the turtle back to 0,0 again
    t.penup()
    t.goto(0,0)
    t.pendown()
    # Set angle to 45 degrees
    t.setheading(45)
    dir = 45
    t.forward(50)

    # Define turtle touched corner
    turtletouchedcorner = False
    endcorner = 'none'
    bounces = 0

    endingcorner = []


    while turtletouchedcorner == False:
        if t.xcor() >= width -20 and t.ycor() >= height -20:
            turtletouchedcorner = True
            endcorner = 'top right'
            # Append endcorner and bounces to endingcorner
            endingcorner.append(endcorner)
            endingcorner.append(bounces)
        if t.xcor() >= width - 20 and t.ycor() <= 20:
            turtletouchedcorner = True
            endcorner = 'bottom right'
            endingcorner.append(endcorner)
            endingcorner.append(bounces)
        if t.xcor() <= 20 and t.ycor() >= height - 20:
            turtletouchedcorner = True
            endcorner = 'top left'
            endingcorner.append(endcorner)
            endingcorner.append(bounces)
        if t.xcor() <= 20 and t.ycor() <= 20:
            turtletouchedcorner = True
            endcorner = 'bottom left'
            endingcorner.append(endcorner)
            endingcorner.append(bounces)
        if t.xcor() >= width and t.heading() == 45:
            t.setheading(135)
            t.forward(1)
            bounces += 1
        elif t.xcor() >= width and t.heading() == 315:
            t.setheading(225)
            t.forward(1)
            bounces += 1
        elif t.xcor() <= 0 and t.heading() == 135:
            t.setheading(45)
            t.forward(1)
            bounces += 1
        elif t.xcor() <= 0 and t.heading() == 225:
            t.setheading(315)
            t.forward(1)
            bounces += 1
        elif t.ycor() >= height and t.heading() == 45:
            t.setheading(315)
            t.forward(1)
            bounces += 1
        elif t.ycor() >= height and t.heading() == 135:
            t.setheading(225)
            t.forward(1)
            bounces += 1
        elif t.ycor() <= 0 and t.heading() == 315:
            t.setheading(45)
            t.forward(1)
            bounces += 1
        elif t.ycor() <= 0 and t.heading() == 225:
            t.setheading(135)
            t.forward(1)
            bounces += 1
        else:
            t.forward(1)

    t.penup()
    t.home()

    height = height / 100
    width = width / 100

    height = int(height)
    width = int(width)

    # Create a table of the results, print them and append save them as a csv file
    results = pd.DataFrame({'Height': [height], 'Width': [width], 'Ending Corner': [endingcorner[0]], 'Bounces': [endingcorner[1]]})
    print(results)
    results.to_csv('results.csv', mode='a', header=False)
t.penup()
```

# Import CSV And sort

Delete duplicate entries

```python
with open('results.csv', 'r') as f:
    lines = f.readlines()
    lines_set = set(lines)
    unique_lines = list(lines_set)
    with open('results.csv', 'w') as f:
        for line in unique_lines:
            f.write(line)
```

```python
import pandas as pd
# Import results.csv and organize it by data in the third column
df = pd.read_csv('results.csv', names=['Height', 'Width', 'Ending Corner', 'Bounces'])
df = df.sort_values(by=['Ending Corner'])
print(df)
```
